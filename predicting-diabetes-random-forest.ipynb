# Let's Now Build Our Random Forest Model


```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
print('Success')
```

Let's declare our variables

```python
y = data.Outcome
features = ['Pregnancies','Glucose','BloodPressure','Insulin','DiabetesPedigreeFunction','Age']
X = data[features]

train_X , val_X, train_y, val_y = train_test_split(X,y, random_state=1)
print('Success')
```

```python
random_forest = RandomForestRegressor(random_state=1)
random_forest.fit(train_X,train_y)

diabetes_predictions = random_forest.predict(val_X)
print(mean_absolute_error(val_y,diabetes_predictions))
```
